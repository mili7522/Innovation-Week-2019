```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
labels = '/media/mike/Files/Data and Results/innovation-challenge-2019/training-labels.csv'
folder = '/media/mike/Files/Data and Results/innovation-challenge-2019/Train/output_combined2'
```

```python
label_df = pd.read_csv(labels)
```

```python
label_df.tail()
```

```python
label_df.info()
```

```python
gp = label_df.groupby('Drscore')
```

```python
gp.count()
```

```python
def show_examples(image_index = 0):
    fig, axes = plt.subplots(3, 2, figsize=(15,15))
    axes = axes.ravel()

    for key, values in gp.groups.items():
        name = label_df.loc[values[image_index], 'Filename']
        file = os.path.join(folder, name)
        data = plt.imread(file)
        axes[key].imshow(data)
        axes[key].axis('off')
        axes[key].set_title(str(key) + ': ' + name)

    axes[-1].axis('off')

show_examples()
```

```python
show_examples(-10)
```
